```python
import re

# Sample phishing email
email = """
From: support-paypal@gmail.com
Subject: Verify Your Account Immediately

Dear User,

Your account has been suspended.
Click here to verify:
http://paypal-security-check.xyz/login

Provide your password immediately.

Thank you.
"""

# Extract sender
sender = re.search(r"From:\s*(.*)", email)
if sender:
    sender = sender.group(1)

# Extract URLs
urls = re.findall(r'https?://[^\s]+', email)

# Phishing indicators
risk_score = 0
indicators = []

if "immediately" in email.lower():
    indicators.append("Urgent Language")
    risk_score += 1

if "password" in email.lower():
    indicators.append("Requests Password")
    risk_score += 1

if sender.endswith("@gmail.com"):
    indicators.append("Suspicious Sender")
    risk_score += 1

for url in urls:
    if ".xyz" in url:
        indicators.append("Suspicious URL")
        risk_score += 1

# Risk Classification
if risk_score >= 3:
    risk = "High"
elif risk_score >= 2:
    risk = "Medium"
else:
    risk = "Low"

# Report
print("=== PHISHING EMAIL ANALYSIS REPORT ===")
print("Sender:", sender)
print("\nURLs Found:")
for url in urls:
    print("-", url)

print("\nIndicators Detected:")
for item in indicators:
    print("-", item)

print("\nRisk Score:", risk_score)
print("Risk Level:", risk)
```
